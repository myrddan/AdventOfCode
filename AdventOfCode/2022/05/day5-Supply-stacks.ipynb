```python
fichier = open("input_challenge.txt", "r")
lignes = fichier.readlines()

cargoInit = [
    ["S","C","V","N"],
    ["Z","M","J","H","N","S"],
    ["M","C","T","G","J","N","D"],
    ["T","D","F","J","W","R","M"],
    ["P","F","H"],
    ["C","T","Z","H","J"],
    ["D","P","R","Q","F","S","L","Z"],
    ["C","S","L","H","D","F","P","W"],
    ["D","S","M","P","F","N","G","Z"]
]

verbose = False
```

---
## Etape 1 Rangement cargo

```python
import re

cargo = [e.copy() for e in cargoInit]

# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = ""


if verbose:
    for col in cargo:
        print(col)

# Pattern pour le regex
pattern = r"move (\d+) from (\d+) to (\d+)"

# Pour chaque instruction des lignes du fichier
for instructions in lignes:

    moveNumber,fromCol, toCol = [int(e) for e in re.search(pattern, instructions).groups()]
    if verbose == True: print(f"{moveNumber=} {fromCol=} {toCol=}")

    for i in range(moveNumber):
        if len(cargo[fromCol-1])>0:
            cargo[toCol-1].append(cargo[fromCol-1].pop())

            if verbose == True:
                for col in cargo:
                    print(col)
                print("")



for col in cargo:
    resultatChallenge+= col.pop()

print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```

## Etape 2 Recherche du nombre de fois ou on passe par 0

```python
import re

cargo = [e.copy() for e in cargoInit]

# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = ""

if verbose:
    for col in cargo:
        print(col)

# Pattern pour le regex
pattern = r"move (\d+) from (\d+) to (\d+)"

# Pour chaque instruction des lignes du fichier
for instructions in lignes:

    moveNumber,fromCol, toCol = [int(e) for e in re.search(pattern, instructions).groups()]
    if verbose == True: print(f"{moveNumber=} {fromCol=} {toCol=}")

    cargo[toCol-1] += cargo[fromCol-1][-int(moveNumber):]
    for i in range(moveNumber):
        cargo[fromCol-1].pop()

    if verbose:
        for col in cargo:
            print(col)

for col in cargo:
    resultatChallenge+= col.pop()

print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
