```python
fichier = open("input_challenge.txt", "r")
lignes = fichier.readlines()

verbose = False
```

---
## Etape 1 Recherche des imbrications totales

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

# Pour chaque instruction des lignes du fichier
for instructions in lignes:
    first,second = instructions.split(",")

    firstStart, firstEnd = first.split("-")
    secondStart, secondEnd = second.split("-")

    firstStart = int(firstStart)
    firstEnd = int(firstEnd)
    secondStart = int(secondStart)
    secondEnd = int(secondEnd)

    if firstStart >= secondStart and firstEnd <= secondEnd:
        resultatChallenge += 1
        if verbose == True: print(f"{firstStart=} >= {secondStart=} and {firstEnd=} <= {secondEnd=}")


    elif firstStart <= secondStart and firstEnd >= secondEnd:
        resultatChallenge += 1
        if verbose == True: print(f"{firstStart=} >= {secondStart=} and {firstEnd=} <= {secondEnd=}")




print(f"\nRésultat du challenge partie 2 : {resultatChallenge}")
```

---
## Etape 2 Recherche des imbrications totales et partielles

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

# Pour chaque instruction des lignes du fichier
for instructions in lignes:
    # On commence par découper les deux plages
    first,second = instructions.split(",")

    # On récupère les bornes des deux plages
    firstStart, firstEnd = first.split("-")
    secondStart, secondEnd = second.split("-")

    # On construit les deux listes de nombres
    first =  [i for i in range(int(firstStart), int(firstEnd) +1)]
    second = [i for i in range(int(secondStart), int(secondEnd) +1)]

    if verbose == True: print(f"{instructions=} > {first=} {second=} ")


    for e in first:
        # Dès que les deux plages on un nombres en comun, on stop la recherche et on incrémente le compteur de résultat
        if e in second:
            resultatChallenge += 1
            break

print(f"\nRésultat du challenge partie 2 : {resultatChallenge}")
```
