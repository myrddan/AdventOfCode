## Utilisation des données du challenges

```python
fichier = open("input_challenge.txt", "r")
lignes = fichier.readlines()

verbose = False
```

---
## Etape 1 Recherche de l'elfe qui a le plus de colorie

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

listeElfes = []
sumElfe = 0

# Pour chaque instruction des lignes du fichier
for calorie in lignes:
    if calorie == "\n":
        if verbose == True: print(f"{sumElfe=}")
        listeElfes.append(sumElfe)
        sumElfe = 0

    else:
        sumElfe = sumElfe+int(calorie)


listeElfes.append(sumElfe)
if verbose == True: print(f"{sumElfe=}")

resultatChallenge = max(listeElfes)

print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```

## Etape 2 Recherche du nombre de fois ou on passe par 0

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

listeElfes = []
sumElfe = 0

# Pour chaque instruction des lignes du fichier
for calorie in lignes:
    if calorie == "\n":
        if verbose == True: print(f"{sumElfe=}")
        listeElfes.append(sumElfe)
        sumElfe = 0

    else:
        sumElfe = sumElfe+int(calorie)


listeElfes.append(sumElfe)
if verbose == True: print(f"{sumElfe=}")

listeElfes.sort()

resultatChallenge = sum(listeElfes[-3:])

print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
