## Etape 1 Calcul du score

Concernant l'opposant ces choix correspondent à :
- A > pierre
- B > feuille
- C > ciseau

Et pour le joueur :
- X > pierre
- Y > feuille
- Z > ciseau



```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

# Pour chaque instruction des lignes du fichier
for match in lignes:
    if verbose == True: print(match.strip())
    opponent,me = match.split()
    scoreManche = 0

    match opponent:
        case "A":
            # Opposant > pierre
            match me:
                case "X":
                    # Moi > pierre
                    scoreManche += 1
                    scoreManche += 3
                    if verbose == True: print(f"  pierre + égalité = 1 + 3")
                case "Y":
                    # Moi > feuille
                    scoreManche += 2
                    scoreManche += 6
                    if verbose == True: print(f"  feuille + gagné = 2 + 6")

                case "Z":
                    # Moi > ciseau
                    scoreManche += 3
                    scoreManche += 0
                    if verbose == True: print(f"  ciseau + perdu = 3 + 0")
        case "B":
            # Opposant > feuille
            match me:
                case "X":
                    # Moi > pierre
                    scoreManche += 1
                    scoreManche += 0
                    if verbose == True: print(f"  pierre + perdu = 1 + 0")
                case "Y":
                    # Moi > feuille
                    scoreManche += 2
                    scoreManche += 3
                    if verbose == True: print(f"  feuille + égalité = 2 + 3")

                case "Z":
                    # Moi > ciseau
                    scoreManche += 3
                    scoreManche += 6
                    if verbose == True: print(f"  ciseau + gagné = 3 + 6")

        case "C":
            # Opposant > ciseau
            match me:
                case "X":
                    # Moi > pierre
                    scoreManche += 1
                    scoreManche += 6
                    if verbose == True: print(f"  pierre + gagné = 1 + 6")
                case "Y":
                    # Moi > feuille
                    scoreManche += 2
                    scoreManche += 0
                    if verbose == True: print(f"  feuille + perdu = 2 + 0")

                case "Z":
                    # Moi > ciseau
                    scoreManche += 3
                    scoreManche += 3
                    if verbose == True: print(f"  ciseau + égalité = 3 + 3")


    resultatChallenge += scoreManche
    if verbose == True: print(f"  {match=} {scoreManche=}")



print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
