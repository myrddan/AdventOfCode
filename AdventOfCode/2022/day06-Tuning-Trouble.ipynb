---
## Etape 1 Recherche du start-of-packet marker

Recherche du marquant de début de packet qui est une suite de caractères différents

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0


# Pour chaque instruction des lignes du fichier
for instructions in lignes:

    for i in range(len(instructions) -3 ):
        extract = instructions[i:i+4]


        isMarker = True
        for c in extract:
            if extract.count(c) > 1:
                isMarker = False
                break

        if isMarker:
            if verbose == True: print(f"{instructions=} {i +4=}")
            resultatChallenge = i +4
            break


print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```

## Etape 2 Recherche du nombre de fois ou on passe par 0

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0


# Pour chaque instruction des lignes du fichier
for instructions in lignes:

    for i in range(len(instructions) -13 ):
        extract = instructions[i:i+14]

        isMarker = True
        for c in extract:
            if extract.count(c) > 1:
                isMarker = False
                break

        if isMarker:
            if verbose == True: print(f"{instructions=} {i +14=}")
            resultatChallenge = i +14
            break


print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
