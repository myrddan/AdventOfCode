## Utilisation des données du challenges

```python
fichier = open("input_challenge.txt", "r")
lignes = fichier.readlines()

verbose = False
```

---
## Etape 1 Recherche du nombre de fois ou on tombe sur 0

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

alphabet="abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"


# Pour chaque instruction des lignes du fichier
for instruction in lignes:
    milieu = len(instruction) // 2

    decoupGauche = instruction[:milieu]
    decoupDroite = instruction[milieu:]

    intersection = list(set(decoupGauche) & set(decoupDroite))

    for letter in intersection:
        resultatChallenge += alphabet.index(letter) +1

    if verbose == True: print(f"{intersection=} {resultatChallenge=}")

print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```

---
## Etape 2 Recherche du nombre de fois ou on passe par 0

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

alphabet="abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Pour chaque instruction des lignes du fichier
for i in range(0,len(lignes),3):

    intersection = list(set(lignes[i].strip()) & set(lignes[i+1].strip()) & set(lignes[i+2].strip()))

    for letter in intersection:
        resultatChallenge += alphabet.index(letter) + 1

    if verbose == True: print(f"{intersection=} {resultatChallenge=}")

print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
