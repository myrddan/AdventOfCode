## Etape 1

```python
resultatChallenge = 0

from itertools import combinations



datas = [ tuple(ligne.strip().split(",")) for ligne in lignes]

max = -1
# generate all combinations of length 2
for j in combinations(datas, 2):
    p1,p2 = j

    x1,y1 = p1
    x2,y2 = p2

    long = abs(int(x1) - int(x2)) +1
    larg = abs(int(y1) - int(y2)) +1

    surface = long * larg

    if surface > max:
        max = surface

    if verbose: print(f"{x1=}:{y1=} >> {x2=}:{y2=} = {surface}")

resultatChallenge = max
print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
