---
## Etape 1 calcul par colonne

```python

# On transforme on liste de str
datas = [ligne.strip().split() for ligne in lignes]
```

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0


NbCalcul = len(datas[0])
NBNumber = len(datas) - 1
# Pour chaque instruction des lignes du fichier
for i in range(NbCalcul):
    chaine = ""
    operation = datas[-1][i]
    for j in range(NBNumber):
        if chaine != "":
            chaine = chaine + operation
        chaine += datas[j][i]

    resultatChallenge += eval(chaine)
    if verbose == True: print(f"{chaine=} = {eval(chaine)}")

print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
