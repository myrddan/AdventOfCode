## Etape 1 Recherche du nombre de fois ou on tombe sur 0

```python
resultatChallenge = 0

# On parcours
for i in range(len(lignes)):
    ligne = lignes[i].strip()

    for j in range(len(ligne)):
        if ligne[j] == "@":
            nbRollOfPaper = 0
            if verbose: print("trouve",ligne[j],">>",i+1,":",j+1)

            #Recherche RollOfPAper ligne au dessus
            if i > 0:
                if lignes[i - 1][j]  == "@":
                    nbRollOfPaper += 1
                    if verbose: print(" ^^ ")

                if j > 0 and lignes[i - 1][j-1]  == "@":
                    nbRollOfPaper += 1
                    if verbose: print(" <^ ")

                if j < len(ligne) -1 and lignes[i - 1][j + 1]  == "@":
                    nbRollOfPaper += 1
                    if verbose: print(" ^> ")

            #Recherche a gauche
            if j > 0:
                if  lignes[i ][j - 1] == "@":
                    nbRollOfPaper += 1
                    if verbose: print(" << ")

            # Recherche a droite
            if j < len(ligne) -1:
                if  lignes[i ][j + 1] == "@":
                    nbRollOfPaper += 1
                    if verbose: print(" >> ")

            #recherche en dessous
            if i < len(lignes) -1 :
                if lignes[i + 1][j]  == "@":
                    nbRollOfPaper += 1
                    if verbose: print(" __ ")

                if j > 0 and lignes[i + 1][j-1]  == "@":
                    nbRollOfPaper += 1
                    if verbose: print(" <_ ")

                if j < len(ligne) -1 and lignes[i + 1][j + 1]  == "@":
                    nbRollOfPaper += 1
                    if verbose: print(" _> ")

            if nbRollOfPaper < 4:
                if verbose: print(f"{nbRollOfPaper=} \n")
                resultatChallenge += 1


print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
