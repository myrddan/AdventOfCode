## Etape 1 Recherche du nombre de rouleau de papier avec moins de quatre rouleaux adjacent

```python
resultatChallenge = 0

# On parcours
for i in range(len(lignes)):
    ligne = lignes[i].strip()

    for j in range(len(ligne)):
        if ligne[j] == "@":
            nbRollOfPaperAdjacent = 0
            if verbose: print("trouve",ligne[j],">>",i+1,":",j+1)

            #Recherche RollOfPAper ligne au dessus
            if i > 0:
                if lignes[i - 1][j]  == "@":
                    nbRollOfPaperAdjacent += 1
                    if verbose: print(" ^^ ")

                if j > 0 and lignes[i - 1][j-1]  == "@":
                    nbRollOfPaperAdjacent += 1
                    if verbose: print(" <^ ")

                if j < len(ligne) -1 and lignes[i - 1][j + 1]  == "@":
                    nbRollOfPaperAdjacent += 1
                    if verbose: print(" ^> ")

            #Recherche a gauche
            if j > 0:
                if  lignes[i ][j - 1] == "@":
                    nbRollOfPaperAdjacent += 1
                    if verbose: print(" << ")

            # Recherche a droite
            if j < len(ligne) -1:
                if  lignes[i ][j + 1] == "@":
                    nbRollOfPaperAdjacent += 1
                    if verbose: print(" >> ")

            #recherche en dessous
            if i < len(lignes) -1 :
                if lignes[i + 1][j]  == "@":
                    nbRollOfPaperAdjacent += 1
                    if verbose: print(" __ ")

                if j > 0 and lignes[i + 1][j-1]  == "@":
                    nbRollOfPaperAdjacent += 1
                    if verbose: print(" <_ ")

                if j < len(ligne) -1 and lignes[i + 1][j + 1]  == "@":
                    nbRollOfPaperAdjacent += 1
                    if verbose: print(" _> ")

            if nbRollOfPaperAdjacent < 4:
                if verbose: print(f"{nbRollOfPaperAdjacent=} \n")
                resultatChallenge += 1


print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```

## Etape 2 Recherche des rouleaux avec moins de 4 adjacents récursif

```python
resultatChallenge = 0

datas = [ligne for ligne in lignes]

while True:

    nbRollOfPaperSup= 0

    # On parcours
    for i in range(len(datas)):
        ligne = datas[i].strip()

        for j in range(len(ligne)):
            if ligne[j] == "@":
                nbRollOfPaperAdjacent = 0
                if verbose: print("trouve",ligne[j],">>",i+1,":",j+1)

                #Recherche RollOfPAper ligne au dessus
                if i > 0:
                    if datas[i - 1][j]  in ["@","#"]:
                        nbRollOfPaperAdjacent += 1
                        if verbose: print(" ^^ ")

                    if j > 0 and datas[i - 1][j-1] in ["@","#"]:
                        nbRollOfPaperAdjacent += 1
                        if verbose: print(" <^ ")

                    if j < len(ligne) -1 and datas[i - 1][j + 1]in ["@","#"]:
                        nbRollOfPaperAdjacent += 1
                        if verbose: print(" ^> ")

                #Recherche a gauche
                if j > 0:
                    if  datas[i ][j - 1] in ["@","#"]:
                        nbRollOfPaperAdjacent += 1
                        if verbose: print(" << ")

                # Recherche a droite
                if j < len(ligne) -1:
                    if  datas[i ][j + 1] in ["@","#"]:
                        nbRollOfPaperAdjacent += 1
                        if verbose: print(" >> ")

                #recherche en dessous
                if i < len(datas) -1 :
                    if datas[i + 1][j] in ["@","#"]:
                        nbRollOfPaperAdjacent += 1
                        if verbose: print(" __ ")

                    if j > 0 and datas[i + 1][j-1] in ["@","#"]:
                        nbRollOfPaperAdjacent += 1
                        if verbose: print(" <_ ")

                    if j < len(ligne) -1 and datas[i + 1][j + 1] in ["@","#"]:
                        nbRollOfPaperAdjacent += 1
                        if verbose: print(" _> ")

                if nbRollOfPaperAdjacent < 4:
                    nbRollOfPaperSup += 1

                    if verbose: print(f"{nbRollOfPaperAdjacent=} {nbRollOfPaperSup=}\n")
                    # Conversion en liste
                    modif = list(datas[i])
                    # Modification du caractère
                    modif[j] = "#"
                    # Reconversion en chaîne
                    datas[i] = "".join(modif)



    for k in range(len(datas)):
        if verbose: print(datas[k].strip())
        modif = list(datas[k])
        for i in range(len(modif)):
            if modif[i] == "#":
                modif[i] = "."
        # Reconversion en chaîne
        datas[k] = "".join(modif)


    if verbose :
        for k in range(len(datas)):
            print(datas[k].strip())

    if nbRollOfPaperSup > 0:
        resultatChallenge += nbRollOfPaperSup

    else:
        break
print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
