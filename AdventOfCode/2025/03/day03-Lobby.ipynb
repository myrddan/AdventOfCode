## Etape 1 Recherche du voltage

You'll need to find the largest possible joltage each bank can produce. In the above example:

- In **98**7654321111111, you can make the largest joltage possible, 98, by turning on the first two batteries.
- In **8**1111111111111**9**, you can make the largest joltage possible by turning on the batteries labeled 8 and 9, producing 89 jolts.
- In 2342342342342**78**, you can make 78 by turning on the last two batteries (marked 7 and 8).
- In 818181**9**1111**2**111, the largest joltage you can produce is 92.


```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

# Pour chaque instruction des lignes du fichier
for banks in lignes:

    banks = banks.strip()
    dizaine : int = 0
    unite : int = 0

    indexDizaine = 0

    # On commence pare recherche la dizaine du voltage
    for i in range(len(banks)-1):
        if int(banks[i]) > dizaine:
            dizaine = int(banks[i])
            indexDizaine = i

    for i in range(indexDizaine + 1,len(banks)):
        if int(banks[i]) > unite:
            unite = int(banks[i])

    joltage = int(str(dizaine) + str(unite))

    if verbose == True: print(f"{banks=} {joltage=}")

    resultatChallenge += joltage
print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```

## Etape 2 Recherche du nombre de fois ou on passe par 0

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

# Pour chaque instruction des lignes du fichier
for banks in lignes:

    banks = banks.strip()
    joltage = ""
    startSearch = 0

    while len(joltage) < 12:

        batterie = 0
        # On commence pare recherche la dizaine du voltage
        for i in range(startSearch, len(banks) - 11+ len(joltage)):
            if int(banks[i]) > batterie:
                batterie = int(banks[i])
                startSearch = i + 1

        joltage += str(batterie)

        if verbose == True: print(f"{banks=} {joltage=}")

    resultatChallenge += int(joltage)
print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
