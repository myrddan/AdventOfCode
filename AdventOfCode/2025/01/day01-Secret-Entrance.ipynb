## Utilisation des données du challenges

```python
fichier = open("input_challenge.txt", "r")
lignes = fichier.readlines()

verbose = False
```

---
## Etape 1 Recherche du nombre de fois ou on tombe sur 0

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

# Stocke la possition du dial entre 0 et 99
posDial = 50

# Pour chaque instruction des lignes du fichier
for instructions in lignes:
    # On récupère la distance du mouvement
    move = instructions[1:]

    # Le premier caractère de l'instruction nous donne le sens
    if instructions[0]== 'L':
        posDial = (posDial - int(move) + 100) % 100
    else:
        posDial = (posDial + int(move)) % 100

    if posDial == 0 :
        resultatChallenge += 1

    if verbose == True: print(f"{instructions=} {posDial=}")

print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```

---
## Etape 2 Recherche du nombre de fois ou on passe par 0

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

posDial = 50

for instructions in lignes:
    # On récupère la distance du mouvement
    move = instructions[1:]

    # On va calculer les tours complets et le reste
    tour, move = divmod(int(move) , 100)

    # Un tours complet ne change rien, que ce soit L ou R on peut l'ajouter directement dans le résultatChallenge
    resultatChallenge += tour

    # Le premier caractère de l'instruction nous donne le sens
    if instructions[0]== 'L':
        # Si je ne suis pas à zéro et que le mouvement va devenir négatif alors je vais passer forcement par zéro.
        if posDial != 0 and posDial - move < 0 : resultatChallenge += 1
        posDial = (posDial - move + 100) % 100
    else:
        # Si je ne suis pas à zéro et que le mouvement est supérieur à 100 alors je vais passer forcement par zéro.
        if posDial != 0 and posDial + move > 100: resultatChallenge += 1
        posDial = (posDial + move) % 100

    # Si j'arrive sur le zéro alors je le compte
    if posDial == 0 : resultatChallenge += 1

    if verbose == True: print(f"{instructions=} {posDial=} {resultatChallenge=}")

print(f"\nRésultat du challenge partie 2 : {resultatChallenge}")
```
