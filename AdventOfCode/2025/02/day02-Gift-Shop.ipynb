## Utilisation des données du challenges

````python
fichier = open("input_challenge.txt", "r")
lignes = fichier.readlines()

verbose = False
````

---
## Etape 1 Recherche des id invalides

Les id invalides sont composé d'un nombre double : 1010 est composé d'un double 10

Il est possible de le faire avec des regex mais on va ici manipuler les str.

````python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

listId = lignes[0].split(",")

# Pour chaque instruction des lignes du fichier
for id in listId:
    min , max = id.split("-")
    if verbose == True: print(f"{min=} {max=}")
    for idChecked in range(int(min), int(max) + 1):
        idStr = str(idChecked)
        milieu = len(idStr) // 2
        if idStr[:milieu] == idStr[milieu:]:
            if verbose == True: print(f"  >> {idChecked=}")
            resultatChallenge += idChecked


print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
````

---
## Etape 2 Recherche des id invalides

Les id invalides sont composé d'un nombre répété plusieurs fois : 1010 est composé d'un double 10,  565656 est composé d'un triple 56

Les regex sont ici la meilleurs solutions

````python
import re

# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

listId = lignes[0].split(",")

# Pour chaque instruction des lignes du fichier
for id in listId:
    min , max = id.split("-")
    min , max = id.split("-")
    if verbose == True: print(f"{min=} {max=}")
    for idChecked in range(int(min), int(max) + 1):
        idStr = str(idChecked)
        milieu = len(idStr) // 2

        pattern= r"\b(\d+)\1+\b"
        if re.match(pattern, idStr):
            if verbose == True: print(f"  >> {idChecked=}")
            resultatChallenge += idChecked

    if verbose == True: print(f"{resultatChallenge=} {resultatChallenge=}")

print(f"\nRésultat du challenge partie 2 : {resultatChallenge}")
````
