## Utilisation des données du challenges

```python
fichier = open("input_challenge.txt", "r")
lignes = fichier.readlines()

verbose = False
```

---
## Etape 1 Calcul du nombre de split

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

data = [ list(ligne.strip()) for ligne in lignes]

for i in range(1,len(data)):
    for j in range(0,len(data[i])):
        if data[i][j] == "." and data[i-1][j] in ["S","|"]:
            data[i][j] = "|"
        elif data[i][j] == "^" and data[i-1][j] == "|":
            data[i][j-1] = "|"
            data[i][j+1] = "|"
            resultatChallenge += 1

if verbose:
    for ligne in data:
        print(ligne)

print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```

---
## Etape 2 Calcul du nombre de chemin possible

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

data = [ list(ligne.strip()) for ligne in lignes]

for i in range(1,len(data)):
    for j in range(0,len(data[i])):
        if data[i][j].isdigit() and  data[i-1][j].isdigit():
            data[i][j] = str(int(data[i][j])+int(data[i-1][j]))
        if data[i][j] == "." and data[i-1][j] =="S":
            data[i][j] = "1"
        elif data[i][j] == "." and data[i-1][j].isdigit():
            data[i][j] = data[i-1][j]
        elif data[i][j] == "^" and data[i-1][j].isdigit():
            if data[i][j-1] == ".":
                data[i][j-1] = data[i-1][j]
            else:
                data[i][j-1] = str(int(data[i][j-1])+int(data[i-1][j]))

            if data[i][j+1] == ".":
                data[i][j+1] = data[i-1][j]
            else:
                data[i][j+1] = str(int(data[i][j+1])+int(data[i-1][j]))


if verbose:
    for ligne in data:
        print(ligne)

for v in data[-1]:
    if v.isdigit():
        resultatChallenge += int(v)
print(f"\nRésultat du challenge partie 2 : {resultatChallenge}")
```
