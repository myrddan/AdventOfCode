---
## Etape 1

```python
# Création d'une variable qui va stocker le résultat du challenge
resultatChallenge = 0

# verbose = True

from itertools import *

# Pour chaque instruction des lignes du fichier
for instructions in lignes:
    decoup=instructions.split()
    lamps = decoup[0].replace('[', '').replace(']', '')
    buttons =  [s.replace("(",'').replace(")",'').split(",") for s in decoup[1:-1]]
    joltage = decoup[-1].replace('{', '').replace('}', '').split(",")

    if verbose: print(f"{lamps=} {buttons=} {joltage=}")

    trouve = False

    for i in range(1,5):
        for result in combinations(buttons, i):
        # for result in combinations_with_replacement(buttons, i):
            if verbose: print(f"{result=}")
            lampsTest = ["." for i in range(len(lamps))]
            if verbose: print("    "+"."*(len(lamps)))
            for button in result:
                if verbose: print(f"    {button=}")
                for v in button:
                    lampsTest[int(v)] = "." if lampsTest[int(v)] == "#" else "#"

                    test = "".join(lampsTest)
                    if verbose: print("   ",test)

            if test == lamps:
                if verbose: print(f"{lamps} >> {result= } {i=}\n\n")
                resultatChallenge+= i
                trouve = True
                break
            if verbose: print("")
        if trouve: break


print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
