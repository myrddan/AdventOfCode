## Etape 2 Recherche du nombre d'ingredient sain de la liste



```python
resultatChallenge = 0
Isingredient = False

freshIds = []
ingredients = []

for ligne in lignes:
    if ligne == "\n":
        Isingredient = True
        continue

    if Isingredient:
        ingredients.append(int(ligne.strip()))
    else:
        start,end = ligne.strip().split("-")
        freshIds.append((int(start),int(end)))


freshIds = list(set(freshIds))
if verbose: print(f"{freshIds=}")
if verbose: print(f"{ingredients=}")


for ingredient in ingredients:
    for start,end in freshIds:
        if ingredient >= start and ingredient <= end:
            resultatChallenge+=1
            break


print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```

---
## Etape 2 Recherche du nombre total des ingredients sains uniques à partir des plages uniquements

```python
resultatChallenge = 0

Isingredient = False

freshIds = []

verbose = False

for ligne in lignes:
    if ligne == "\n":
        break

    if verbose: print(f"{ligne=}")
    start,end = ligne.strip().split("-")
    freshIds.append([int(start),int(end)])

if verbose: print(f"{freshIds=}")

finit = False

verbose = True

while not finit:
    finit = True

    for  i in range(1,len(freshIds)):
        sup = False

        for y in range(0,i+1):

            if i == y :continue

            start1,end1 = freshIds[y]
            start2,end2 = freshIds[i]

            if start2 <= start1 <= end2:
                start1 = start2
                sup = True

            if start2 <= end1 <= end2:
                end1 = end2
                sup = True

            if sup:
                if verbose: print(f"update {freshIds[y]} >> {[start1,end1]}")
                freshIds[y] = [start1,end1]

                if verbose: print(f"delete {freshIds[i]}")
                del freshIds[i]
                break


            if start1 <= start2 <= end1:
                start2 = start1
                sup = True

            if start1 <= end2 <= end1:
                end2 = end1
                sup = True

            if sup:
                if verbose: print(f"update {freshIds[i]} >> {[start2,end2]}")
                freshIds[i] = [start2,end2]

                if verbose: print(f"delete {freshIds[y]}")
                del freshIds[y]
                break


        if sup:
            finit = False
            break

verbose = True

if verbose: print(f"")

for plage in freshIds:
    if verbose: print(f"{plage=} >> nombre {plage[1]-plage[0]+1}")
    resultatChallenge += plage[1]-plage[0]+1


print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
print(f"{resultatChallenge-366181852921027=}")
```
