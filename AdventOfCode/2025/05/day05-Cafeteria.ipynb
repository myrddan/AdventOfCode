## Etape 1 : Calculer le nombre d'ingredient encore bon

J'ai voulu sans regarder les vrai donnéés, convertir les plages en une seul list d'ingrédient sain ce qui facilite le calcul.

Mais vu la tailles des nombres de la plages. le code est trop gourmant

```python
Isingredient = False;

freshIds = []
ingredients = []

for ligne in lignes:
    if ligne == "\n":
        Isingredient = True
        continue

    if Isingredient:
        ingredients.append(int(ligne.strip()))
    else:
        if verbose: print(f"{ligne=}")
        start,end = ligne.strip().split("-")

        freshIds += [i for i in range(int(start), int(end) + 1)]

freshIds = list(set(freshIds))
if verbose: print(f"{freshIds=}")
if verbose: print(f"{ingredients=}")

resultatChallenge = 0

for ingredient in ingredients:
    if ingredient in freshIds:
        resultatChallenge+=1


print(f"\nRésultat du challenge partie 1 : {resultatChallenge}")
```
